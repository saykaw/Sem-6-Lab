```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Data-Based D3.js Pie Chart Example</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>

    <svg id="pieChart" width="400" height="400"></svg>

    <script>
        var data = [30, 40, 50];

        var svg = d3.select("#pieChart")
            .append("g")
            .attr("transform", "translate(200,200)"); // Center the pie chart

        var pie = d3.pie();

        var arc = d3.arc()
            .innerRadius(0)
            .outerRadius(100);

        var arcs = svg.selectAll("path")
            .data(pie(data))
            .enter().append("path")
            .attr("d", arc)
            .attr("fill", function(d, i) {
                return d3.schemeCategory10[i]; // Use a color scheme for each slice
            })
            .on("mouseover", function() {
                d3.select(this)
                    .transition()
                    .attr("d", arc.innerRadius(0).outerRadius(120)); // Enlarge the selected slice
            })
            .on("mouseout", function() {
                d3.select(this)
                    .transition()
                    .attr("d", arc.innerRadius(0).outerRadius(100)); // Reset the size
            });
    </script>

</body>
</html>
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3.js Example</title>
    <style>
        .highlight {
            color: red;
        }
        body {
            font-family: Arial, sans-serif;
        }
    </style>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>

    <!--Selecting and Modifying Elements-->
    <p id="myParagraph">This is a paragraph.</p>

    <!--Buttons for Dynamic Modifications-->
    <button id="changeTextButton">Change Text</button>
    <button id="addParagraphButton">Add Paragraph</button>

    <script>
        //Selecting Elements
        var paragraphElement = d3.select("#myParagraph");
        var allParagraphs = d3.selectAll("p");

        //Modifying Elements
        paragraphElement.text("Original Text :(");
        paragraphElement.attr("class", "highlight");
        paragraphElement.style("color", "blue");
        d3.select("body").append("div").text("New Div Element");

        d3.select("#changeTextButton").on("click", function() {
            paragraphElement.text("Updated Text on Button Click :)");
        });

        d3.select("#addParagraphButton").on("click", function() {
            d3.select("body").append("p").text("New Text Added!");
        });

    </script>

</body>
</html>

```
