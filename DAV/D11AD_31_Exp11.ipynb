##Visualization in Python

```r
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import folium
```

```r
df = pd.read_csv("StudentsPerformance.csv")
```

```r
df.head()
```

```r
df.describe()
```

```r
df.info()
```

```r
df.isnull().sum()       #no null values
```

##Categorical Variables

```r
sns.countplot(x='gender', data=df)
plt.title('Distribution of Gender')
plt.show()
```

```r
sns.countplot(x='race/ethnicity', data=df)
plt.title('Distribution of Race/Ethnicity')
plt.show()
```

```r
lunch_counts = df['lunch'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue'])
plt.title('Distribution of Lunch Type')
plt.show()
```

##Numeric Variables

```r
plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['math score'], bins=30, kde=True, color="skyBlue")
plt.title('Distribution of Math Scores')
```

```r
plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], bins=30, kde=True, color = "plum")
plt.title('Distribution of Reading Scores')
```

```r
plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], bins=30, kde=True, color ="green")
plt.title('Distribution of Writing Scores')
```

```r
import plotly.express as px
correlation_matrix = df[['math score', 'reading score', 'writing score']].corr()
fig = px.imshow(correlation_matrix, color_continuous_scale='Viridis')
fig.update_layout(title='Correlation Heatmap of Scores', xaxis_title='Scores', yaxis_title='Scores')
fig.show()
```

**We can see that reading score and writing score are closely related**

##Visualization in R

```r
library(dplyr)
library(ggplot2)
```

```r
df <- read.csv("StudentsPerformance.csv")
```

```r
#scatter plot of math score vs. reading scores, colored by gender
options(repr.plot.width = 10, repr.plot.height = 6)
ggplot(df, aes(x = math.score, y = reading.score, color = gender)) +
  geom_point() +
  labs(title = "Scatter Plot of Math vs. Reading Scores", x = "Math Score", y = "Reading Score") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```r
# Stacked bar plot for 'race/ethnicity' and 'gender'
ggplot(df, aes(x = race.ethnicity, fill = gender)) +
  geom_bar(position = "stack") +
  labs(title = "Stacked Bar Plot of Race/Ethnicity by Gender", x = "Race/Ethnicity", y = "Count") +
   theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```
